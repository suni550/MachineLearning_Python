```python
import numpy as np
from PIL import Image

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
img = Image.open('dogs.jpeg')
img
```

```python
plt.imshow(img)
```

```python
np_img = np.asarray(img)
np_img.shape
```

##### K-Means algo expects the 2-D array as input. But the 'np_img' is 3-D image. So reshape 3-D into 2-D

```python
np_img = np_img.reshape(185*272, 3)
```

```python
np_img.shape
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(np_img)
```

```python
predicted = kmeans.predict(np_img)
predicted = predicted.reshape(185, 272)
predicted.shape
plt.imshow(predicted)
```
